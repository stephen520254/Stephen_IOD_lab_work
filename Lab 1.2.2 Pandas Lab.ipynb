# 5. Visualization

> Indented block



#### 5.A Plot Histogram of Age 

````python
# A:
x= df.Age.tolist()
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.hist(x,bins=10);
plt.xlabel('Age');
plt.ylabel('Count');
plt.title('Age Profile');
````

#### 5.B BoxPlot of Age

````python
# A:
fig=plt.gcf()
fig.set_size_inches(10,6)
sns.boxplot(x);
plt.xlabel('Age')
plt.title('Boxplot of Age')
````

#### 5.C Scatterplot of age and length_of_service

````python
# A:
fig=plt.gcf()
fig.set_size_inches(10,6)
x=df.Age
y=df.YearsInCurrentRole

plt.scatter(x,y, s=1);
plt.xlabel('Age');
plt.ylabel('Years In Current Role')
plt.title('Scatter plot of Years in current role by Age')
````

#### 5.D Plot Histogram of store_name

````python
# A:

fig=plt.gcf()
fig.set_size_inches(10,6)
x=df.HourlyRate
plt.hist(x, bins=10);
plt.xlabel('Hourly Rate');
plt.ylabel('Counts');
plt.title('Hourly Role');
````
