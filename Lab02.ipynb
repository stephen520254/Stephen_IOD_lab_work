```python
# Practice the use of matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
# Create a data set 
x = np.linspace(0, 10, 50)
# Print the data
print(x)
# Create a figure using plt.figure
fig = plt.figure(figsize=(8,8))
# Plot the sine wave of x
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '-')
plt.show
```
